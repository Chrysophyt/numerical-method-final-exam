```
from sympy import symbols, diff, solve, ln
import pandas as pd
import numpy as np
import math as mt
pd.options.display.float_format = '{:,.2f}'.format
```

# Newton's Method on Non-linear system
more info at : 
http://www.ohiouniversityfaculty.com/youngt/IntNumMeth/lecture13.pdf

```
x, y = symbols('x y', real=True)

listSymbols = [x, y]
f = []
f.append(x*y-mt.sqrt(8))
f.append(x**2 - y**2-2)

#creating Df
Df = [[None]*len(listSymbols) for i in range(len(f))]
for i in range(len(f)):
  for j in range(len(listSymbols)):
    Df[i][j] = diff(f[i], listSymbols[j])
print("Df : ",Df)
```

```
# iterate
x0, y0 = 10, 0
n = 10

Dfval = [[None]*len(listSymbols) for i in range(len(f))]
Dfi = []
deltai = []
fval = [None]*len(f)
fi = []
xi = []
yi = []
xi.append(x0)
yi.append(y0)

# Df(xi, yi) * delta =- f(xi, yi)
for k in range(n):
  #set xk -> Df
  for i in range(len(f)):
    for j in range(len(listSymbols)):
      Dfval[i][j] = Df[i][j].subs([(x, xi[k]), (y, yi[k])])
  dfi = np.array(Dfval, np.float32)
  Dfi.append(dfi)
  
  #set xk -> f
  for i in range(len(f)):
    fval[i] = f[i].subs([(x, xi[k]), (y, yi[k])])
  fvalue = np.negative(np.array(fval, np.float32))
  fi.append(fvalue)

  Delta = np.linalg.solve(Dfi[k], fi[k])
  deltai.append(Delta)
  xi.append(xi[k]+deltai[k][0])
  yi.append(yi[k]+deltai[k][1])


    
```

```
d = {'xi':xi, 'yi':yi, 'Dfi':Dfi, 'fi':fi, 'deltai':deltai}
df = pd.DataFrame.from_dict(d, orient='index')
df = df.transpose()
df.index.name = 'i'
df
```
