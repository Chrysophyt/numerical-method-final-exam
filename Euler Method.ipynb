```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as mt
pd.options.display.float_format = '{:,.2f}'.format
```

```
# Manual :(
def diff(x, y):
  return 2*x-y+1

def y_analytical(x):
  return 2*x+2*mt.e**(-x)-1
```

## Using Euler Method

```
ArrayY = []
ArrayX = []
ArrayF = []

X_MAX = 2
iX, iY = 0, 1
stepSize = 0.25

ArrayX.append(iX)
ArrayY.append(iY)
ArrayF.append(0)

for i in range(int(X_MAX/stepSize)):
    iX+=stepSize
    ArrayX.append(iX)
    ArrayF.append(diff(ArrayX[i], ArrayY[i]))
    ArrayY.append(ArrayY[i]+ diff(ArrayX[i], ArrayY[i])*stepSize)


data = {'xi':ArrayX, 'yi':ArrayY, 'f(xi,yi)':ArrayF}
df = pd.DataFrame(data)
df
```

## Plot graph

```
# Plot Euler
for i in range(0, len(ArrayX)):
    plt.plot(ArrayX[i:i+2],ArrayY[i:i+2],'ok-')

# Plot Analytical
x_vals = np.linspace(0, 2, 100)
y_vals = y_analytical(x_vals)
plt.plot(x_vals, y_vals)
plt.show()
```
